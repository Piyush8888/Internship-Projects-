```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame with columns: age, sex, bmi, children, smoker, region, charges
# You may need to replace 'data' with the actual variable name you are using.

# Bar chart for gender distribution
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', data=data, palette='Set2')
plt.title('Distribution of Genders')
plt.show()

# Histogram for age distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['age'], bins=30, kde=False, color='skyblue')
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Assuming 'data' is your DataFrame with columns: age, sex, smoker
# You may need to replace 'data' with the actual variable name you are using.

# Filter data for smokers only
smokers_data = data[data['smoker'] == 'yes']

# Create age groups (you can customize the bins based on your data)
age_bins = [18, 30, 40, 50, 60, 70, 80, 90]
labels = ['18-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89']
smokers_data['age_group'] = pd.cut(smokers_data['age'], bins=age_bins, labels=labels, right=False)

# Create a bar plot to compare age groups of male and female smokers
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.countplot(x='age_group', hue='sex', data=smokers_data, palette='pastel')
plt.title('Age Group Distribution of Smokers by Gender')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
# Filter data for smokers only
smokers_data = data[data['smoker'] == 'yes']

# Create a boxplot to visualize charges by region for smokers
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))
sns.boxplot(x='region', y='charges', data=smokers_data, palette='Set2')
plt.title('Charges Paid by Smokers in Different Regions')
plt.xlabel('Region')
plt.ylabel('Charges')
plt.show()
```

```python
# Filter data for smokers only
smokers_data = data[data['smoker'] == 'yes']

# Group data by region and calculate the sum of charges
charges_by_region = smokers_data.groupby('region')['charges'].sum()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(charges_by_region, labels=charges_by_region.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)
plt.title('Distribution of Charges by Region for Smokers')
plt.show()
```

* This pie chart provides a visual representation of the proportion of charges contributed by each region among smokers.

```python
plt.figure(figsize=(12, 8))
sns.histplot(data, x='charges', hue='sex', multiple='stack', bins=30, palette='Set2', element='step', common_norm=False)
plt.title('Distribution of Charges by Sex and Region')
plt.xlabel('Charges')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.stats import zscore
# Calculate Z-scores for the 'age' column
data['age_zscore'] = zscore(data['age'])

# Set a threshold for Z-scores (e.g., 3 for a significance level of 99.7%)
zscore_threshold = 3

# Identify outliers based on Z-scores
outliers = data[data['age_zscore'].abs() > zscore_threshold]

# Plot the age distribution with outliers highlighted
plt.figure(figsize=(10, 6))
plt.hist(data['age'], bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.scatter(outliers['age'], [0] * len(outliers), color='red', marker='x', label='Outliers')
plt.title('Age Distribution with Outliers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

* This visualization helps you identify data points that deviate significantly from the mean in the age distribution. 

```python
# Create a scatter plot to visualize the relationship between age and charges
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='charges', data=data, color='skyblue')
plt.title('Scatter Plot: Age vs Charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()

# Calculate the correlation coefficient
correlation = data['age'].corr(data['charges'])
print(f"Correlation between age and charges: {correlation}")
```

* The x-axis represents age, and the y-axis represents charges.
* The correlation coefficient is calculated using the corr method.
* The scatter plot visually displays the points and their distribution, and the correlation coefficient quantifies the strength and direction of the linear relationship between age and charges.

```python
# Calculate the count of smokers and non-smokers
smoker_counts = data['smoker'].value_counts()

# Calculate the proportion of smokers vs. non-smokers
smoker_proportion = smoker_counts / len(data)

# Print the results
print("Smoker Counts:")
print(smoker_counts)

print("\nProportion of Smokers vs. Non-Smokers:")
print(smoker_proportion)
```
